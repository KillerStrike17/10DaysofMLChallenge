## Importing heavenly Bodies

```python
#time to call the gods
import pandas as pd
from sklearn import preprocessing
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

## Data Loading

```python
def preprocess_df(df):
    processed_df = df.copy()
    le = preprocessing.LabelEncoder()
    processed_df.Sex = le.fit_transform(processed_df.Sex)
    processed_df.Embarked = le.fit_transform(processed_df.Embarked)
    processed_df.Name = le.fit_transform(processed_df.Name)
    processed_df.Ticket = le.fit_transform(processed_df.Ticket)
    return processed_df
```

```python
dataset=pd.read_csv("http://iali.in/datasets/titanic/train.csv")
```

```python
dataset.head()
```

## Data Visualization

```python
f,ax=plt.subplots(2,2,figsize=(10,6))
sns.countplot(dataset['Survived'],ax=ax[0,0])
dataset['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',shadow=True, startangle=90,ax = ax[0,1])
ax[0,1].set_title('Survived')
ax[0,1].set_ylabel('')
dataset[['Sex','Survived']].groupby(['Sex']).mean().plot.bar(ax = ax[1,0])
sns.countplot('Sex',hue='Survived',data=dataset,ax = ax[1,1])
ax[1,1].set_title('Sex:Survived vs Dead')
```

From the graphs we can see that women survived more, out of the total women , majority women survived

```python
f,ax=plt.subplots(2,2,figsize=(10,10))
sns.countplot(dataset['Embarked'],ax=ax[0,0])
ax[0,0].set_title('No. Of Passengers Boarded')
sns.countplot(dataset['Embarked'],hue=dataset['Sex'],ax = ax[0,1])
ax[0,1].set_title('Male-Female Split for Embarked')
sns.countplot(dataset['Embarked'],hue=dataset['Survived'],ax=ax[1,0])
ax[1,0].set_title('Embarked vs Survived')
sns.countplot(dataset['Embarked'],hue=dataset['Pclass'],ax=ax[1,1])
ax[1,1].set_title('Embarked vs Pclass')
plt.show()
```

```python
f,ax=plt.subplots(figsize=(5,5))
sns.barplot(dataset['SibSp'],dataset['Survived'],ax=ax)
ax.set_title('SibSp vs Survived')
plt.show()
```

SibSp having 1 sibling had more chances of survival

```python
dataset = dataset.drop(columns = ['Cabin'])
dataset.head()
```

 We will check where we need to impute the data or drop (Age column). Based on the correlation of it with the final variable.To do that we need to find the correlation of not Nan dataset.
 

```python
nn_dataset = dataset.dropna()
```

```python
nn_dataset.head()
```

We need to convert the text to numerical format before finding correlations

```python
processed_df = preprocess_df(nn_dataset)
# print(processed_df)
processed_df.head()
```

```python
corr = processed_df.corr()
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns,annot=True,cmap='RdYlGn')
```

Here we can see that, there is no correation at all of columns Age, PassengerID, Name, SibSp & Parch. So we can drop those

```python
processed_df = processed_df.drop(columns = ['Age', 'PassengerId', 'Name', 'SibSp', 'Parch'])
```
